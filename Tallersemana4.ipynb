```python
import numpy as np
import matplotlib.pyplot as plt
import random

def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib

def graficar_fibonacci(n):
    serie = fibonacci(n)
    plt.figure(figsize=(8, 5))
    plt.plot(range(n), serie, marker='o', linestyle='-', color='b', label='Serie de Fibonacci')
    plt.xlabel('Índice')
    plt.ylabel('Valor')
    plt.title('Serie de Fibonacci')
    plt.legend()
    plt.grid()
    plt.show(block=True)
def analisis_temperaturas():
    temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]
    media = np.mean(temperaturas)
    mediana = np.median(temperaturas)
    desviacion = np.std(temperaturas)
    print(f"Media: {media}, Mediana: {mediana}, Desviación estándar: {desviacion}")
    plt.plot(temperaturas, marker='o', linestyle='-', color='r')
    plt.title('Temperatura media mensual')
    plt.xlabel('Mes')
    plt.ylabel('Temperatura (°C)')
    plt.grid()
    plt.show(block=True)


def simulacion_dado():
    resultados = [random.randint(1, 6) for _ in range(1000)]
    plt.hist(resultados, bins=np.arange(1, 8)-0.5, edgecolor='black', rwidth=0.8)
    plt.xlabel('Número del dado')
    plt.ylabel('Frecuencia')
    plt.title('Simulación de 1000 lanzamientos de un dado')
    plt.xticks(range(1, 7))
    plt.grid()
    plt.show(block=True)

def graficar_trigonometricas():
    x = np.linspace(0, 2*np.pi, 100)
    plt.plot(x, np.sin(x), label='sin(x)')
    plt.plot(x, np.cos(x), label='cos(x)')
    plt.plot(x, np.tan(x), label='tan(x)')
    plt.ylim(-2, 2)
    plt.legend()
    plt.title('Funciones Trigonométricas')
    plt.grid()
    plt.show(block=True)

def graficar_raices():
    x = np.linspace(0, 10, 100)
    plt.plot(x, np.sqrt(x), label='y = sqrt(x)')
    plt.plot(x, x, label='y = x', linestyle='--')
    plt.legend()
    plt.title('Raíces Cuadradas')
    plt.grid()
    plt.show(block=True)

def pendiente_recta(x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)
    print(f'Pendiente de la recta: {m}')
    x = np.array([x1, x2])
    y = np.array([y1, y2])
    plt.plot(x, y, marker='o', linestyle='-', label='Recta')
    plt.legend()
    plt.grid()
    plt.title('Recta entre dos puntos')
    plt.show(block=True)

def movimiento_parabolico(v0, theta):
    g = 9.81
    theta = np.radians(theta)
    t_max = (2 * v0 * np.sin(theta)) / g
    t = np.linspace(0, t_max, num=100)
    x = v0 * np.cos(theta) * t
    y = v0 * np.sin(theta) * t - 0.5 * g * t**2
    plt.plot(x, y, label='Trayectoria')
    plt.xlabel('Distancia (m)')
    plt.ylabel('Altura (m)')
    plt.legend()
    plt.grid()
    plt.title('Movimiento Parabólico')
    plt.show(block=True)




graficar_fibonacci(10)
analisis_temperaturas()
simulacion_dado()
graficar_trigonometricas()
graficar_raices()
pendiente_recta(1, 2, 4, 6)
movimiento_parabolico(10, 45)
```
